# **1.4.3 Matplotlib**

Graficar no sólo es importante para papers, libros, es sumamenta útil pata darle un vistazo rápido a los datos y puede ahorrar mucho tiempo cuando se trabaja con bases de datos grandes.

Permite crear y personalizar los tipos de gráficos más comunes, entre ellos:

* Diagramas de líneas
* Diagramas de barras
* Diagramas de dispersión
* DIagrama de área
* Mapas de color

y combinaciones entre ellas. Para más información: https://matplotlib.org/stable/plot_types/index



En el ejemplo siguiente la matriz se ve como:

\begin{equation}
\begin{matrix}
-2 & -1 & 0 & 1 & 2\\
-4 & -2 & 0 & 2 & 4
\end{matrix}
\end{equation}

```python
%matplotlib inline
import matplotlib.pyplot as plt

matriz = [[-2, -1, 0, 1, 2],[-4, -2, 0, 2, 4]]

x_vals = matriz[0][:]
y_vals = matriz[1][:]
plt.plot(x_vals, y_vals)
plt.title("Diagrama de líneas")
plt.ylabel("Eje y")
plt.xlabel("Eje x");
```

Es posible especificar tamaño y color de los elementos que conforman el plot

**matplotlib.pyplot.scatter(x, y, s=None, c=None, marker=None, alpha=None, linewidths=None, *, edgecolors=None)**

```python
transparency = 0.6
size = 40

plt.scatter(x_vals, y_vals, s=size, c='red', edgecolors='none', alpha=transparency)
plt.title("Diagrama de dispersión")
plt.ylabel("Eje y")
plt.xlabel("Eje x")
```

En el siguiente ejemplo los valores de x y y se ven así

\begin{equation}
\begin{matrix}
x: & 1 & 2 & 3 & 4\\
y: & 1 & 2 & 0 & 0.5
\end{matrix}
\end{equation}

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.fill_between([1, 2, 3, 4], [1, 2, 0, 0.5])
plt.title("Diagrama de área")
plt.ylabel("Eje y")
plt.xlabel("Eje x")
plt.show()
```

En el siguiente ejemplo los valores de x y y se ven así

\begin{equation}
\begin{matrix}
x: & 1 & 2 & 3 \\
y: & 3 & 2 & 1 
\end{matrix}
\end{equation}

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.bar([1, 2, 3], [3, 2, 1])
plt.title("Diagrama barras")
plt.ylabel("Eje y")
plt.xlabel("Eje x")
plt.show()
```
