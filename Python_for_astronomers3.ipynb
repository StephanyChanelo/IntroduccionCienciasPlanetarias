**Imagen**

Una imagen se define como una función de dos dimensiones f(x,y) donde x e y son las coordenadas de un plano que contiene todos los puntos de la misma, y f(x,y) es la amplitud en el punto (x,y) a la cual se le llama intensidad o nivel de gris de la imagen en ese punto. 

Una imagen digital está compuesta de un número finito de elementos y cada uno tiene una localidad y un valor particular. A estos elementos se les llama puntos elementales de la imagen o **píxeles**, siendo este último el término comúnmente utilizado para denotar la unidad mínima de medida de una imagen digital. 

![alt text](https://miro.medium.com/max/693/1*bV7S0zACdidh11ikjYpLpQ.png)


**Imagen de color** \\
El fundamento para describir una imagen digital en color es similar con la diferencia de que cada píxel es descrito y codificado de otra forma, según el espacio de color que se esté utilizando. Así por ejemplo, para un espacio de color RGB (generalmente el más usado para representar imágenes), se representa cada píxel como un color creado a partir de ciertas cantidades de los colores rojo, verde y azul. Esta representación se puede interpretar como una matriz de tres niveles de intensidad, donde cada nivel corresponde a la intensidad de color de las componentes rojo, verde y azul.



![alt text](https://www.researchgate.net/profile/Bhupendra_Pratap_Singh/publication/282798184/figure/fig15/AS:283989639221249@1444719814399/Color-image-representation-and-RGB-matrix.png)

Para la comunicación científica, los dos espacios de color principales son RGB y CMYK.

**RGB**

El modelo de color RGB se relaciona muy de cerca con la forma en que percibimos el color con los receptores r, g y b en nuestras retinas. RGB usa una mezcla de color aditiva y es el modelo de color básico usado en televisión o cualquier otro medio que proyecte color con luz.

rgbm.png


**Entonces como creamos una imagen astronómica en RGB?**

Es posible crear imágenes en color cercanas al “color verdadero” si existen tres imagenes con filtros cercanos a los receptores r, g y b de nuestros ojos. Las imágenes que se aproximan a lo que vería un viajero espacial ficticio si realmente viajara al objeto se denominan imágenes de “color natural”.

Para hacer una imagen de color natural, el orden de los colores asignados a las diferentes exposiciones debe ser en "orden cromático", es decir, la longitud de onda más baja debe tener un tono azul, la longitud de onda media debe tener un tono verde y la longitud de onda más alta debe ser roja.

**Histograma**

Supongamos dada una imagen en niveles de grises, siendo el rango de
256 tonos de gris (de 0 a 255). El histograma es un diagrama de barras donde los valores en x representan los niveles de gris de una imagen, y los valores en y las frecuencias relativas de los distintos niveles de gris, es decir, la cantidad de puntos asociados a cada nivel de gris, dividido por la cantidad total de puntos de la imagen.


\begin{equation}
P_{k} = \frac{n_{k}}{n}
\end{equation}


$n_{k}$ = cantidad de pixels de nivel gris

$n$ = cantidad total de pixeles


```python
hist_r = cv2.calcHist(r, [0], None, [256], [0, 256])
hist_g = cv2.calcHist(g, [0], None, [256], [0, 256])
hist_b = cv2.calcHist(b, [0], None, [256], [0, 256])

#plt.subplot(1,3,1), 
plt.plot(hist_r),plt.xlim([0,256]),plt.ylim([0,25])
#plt.subplot(1,3,2), plt.plot(hist_g),plt.xlim([0,256]),plt.ylim([0,25])
#plt.subplot(1,3,3), plt.plot(hist_b),plt.xlim([0,256]),plt.ylim([0,25])

plt.show()
```

En la grafica anterior se observa un histograma estrecho, de bajo rango dinámico, en concordancia con el poco contraste visual de la imagen correspondiente. Es decir, una imagen oscura, ya que la mayoria de sus pixeles se encuentran concentrados en valores bajos dde intensidad [0-100]. 

**Ecualización del histograma**

Cuando el rango de niveles de gris que toma la imagen se
encuentra concentrado en una zona del intervalo, la imagen posee
poco contraste. Para aumentar el contraste, podemos expandir el
histograma o bien realizar una ecualización del mismo.

El objetivo principal del histograma es ampliar el rango dinámico de los niveles de gris, para acentuar el contraste visual de la imagen. Es decir, el histograma podría ser alterado para producir cambios en la imagen. La ecualización mejora el contraste de la imagen. Reparte de forma más o menos uniforme los valores del histograma.

```python
equ_r = cv2.equalizeHist(r)
equ_g = cv2.equalizeHist(g)
equ_b = cv2.equalizeHist(b)
hist_equr = cv2.calcHist(equ_r, [0], None, [256], [0, 256])
plt.plot(hist_equr),plt.xlim([0,256]),plt.ylim([0,25])
plt.show()
```

```python
figure_size = 15
plt.figure(figsize=(figure_size,figure_size))
plt.subplot(1,3,1),plt.imshow(r, cmap='gray'), plt.title("R")
plt.subplot(1,3,2),plt.imshow(equ_r, cmap='gray'), plt.title("R - ecualizado")
plt.show()
```

```python
color_image = cv2.merge((b,g,r))
color_equimage = cv2.merge((equ_b,equ_g,equ_r))

figure_size = 15
plt.figure(figsize=(figure_size,figure_size))
plt.subplot(1,3,1),plt.imshow(color_image), plt.title("Imagen a color")
plt.subplot(1,3,2),plt.imshow(color_equimage), plt.title("Imagen ecualizada")
plt.show()
```

**Realce de imagen**

La técnica de realce es fundamental en el procesamiento de imágenes ya que permite mejorar las características de brillo y por consiguiente la calidad visual de una imagen. La transformación aplicada a la imagen de entrada se puede hacer píxel a píxel o a un grupo de píxeles. El proceso altera el nivel de gris de los píxeles de la imagen mejorando así la percepción visual de los detalles en la imagen. 

Estas técnicas además de mejorar la apariencia visual de la imagen también son fundamentales para el manejo del ruido. La transformación analizada e implementada en este articulo corresponde a la transformación lineal por medio de rectas. Consiste en aplicar una función matemática que corresponde a la de una recta, sobre la imagen que se desea procesar, y más puntualmente en una vecindad de píxeles, por lo tanto se implementó una función por tramos. El proceso ocasiona en la imagen la contracción o estiramiento de contraste, dependiendo de la pendiente de la recta de transformación; en el caso del estiramiento la diferencia en el contraste entre píxeles se hace más notoria.

\begin{equation}
Y = α x + β
\end{equation}


```python
alpha=1.5
beta=5
new_image = cv2.convertScaleAbs(color_image, alpha=alpha, beta=beta)

figure_size = 15
plt.figure(figsize=(figure_size,figure_size))
plt.subplot(1,3,1),plt.imshow(color_image), plt.title("Imagen a color")
plt.subplot(1,3,2),plt.imshow(new_image), plt.title("Contraste e iluminación modificado")

plt.show()
```

**Filtros**

Los filtros digitales constituyen uno de los principales modos
de operar en el procesamiento de imágenes digitales. Pueden
usarse para distintos fines, pero en todos los casos, el
resultado sobre cada píxel depende de los píxeles de su
entorno. Los principales objetivos:



1.   **Suavizar la imagen:** reducir las variaciones de intensidad
entre píxeles vecinos.
2.   **Eliminar ruido:** modificar aquellos píxeles cuyo nivel de
intensidad es muy diferente al de sus vecinos.
3. **Realzar la imagen:** aumentar las variaciones de intensidad,
allí donde se producen.
4. **Detectar bordes:** detectar aquellos píxeles donde se produce
un cambio brusco en la función intensidad.

**Qué es un kernel?**

Un kernel es esencialmente una matriz de tamaño fijo de coeficientes numéricos junto con un punto de anclaje en esa matriz, que normalmente se encuentra en el centro.

![alt text](https://docs.opencv.org/4.x/filter_2d_tutorial_kernel_theory.png)

**Filtro para suavizar la imagen**

**Filtro de la media**

El filtro de la media es el más simple, intuitivo y fácil de implementar
para suavizar imágenes que el de la mediana, es decir, reducir la
cantidad de variaciones de intensidad entre píxeles vecinos.

Se visita cada píxel de la imagen y se reemplaza
por la media de los píxeles vecinos.


![alt text](https://3.bp.blogspot.com/-z1EIsBG_8hA/U6KlbsJoBMI/AAAAAAAABdQ/jgThabZzWDM/s1600/avg.PNG)

```python
kernel_3x3 = np.ones((15, 15), np.float32) / 225
blurred = cv2.filter2D(color_image, -1, kernel_3x3)

figure_size = 15
plt.figure(figsize=(figure_size,figure_size))
plt.subplot(1,3,1),plt.imshow(color_image), plt.title("Imagen original")
plt.subplot(1,3,2),plt.imshow(blurred), plt.title("Imagen suavizada")

plt.show()
```

**Filtro de nitidez**

El filtro de nitidez en el procesamiento de imágenes mejora la resolución espacial al mejorar los límites de los objetos, pero a costa del ruido de la imagen:

1. Destacar los detalles finos.
2. Mejorar los detalles que se han desenfocado.


![alt text](https://www.globalsino.com/EM/image2/1371a.GIF)

```python
kernel_sharpening = np.array([[-1,-1,-1], 
                              [-1,9,-1], 
                              [-1,-1,-1]])
sharpened = cv2.filter2D(color_image, -1, kernel_sharpening)

figure_size = 15
plt.figure(figsize=(figure_size,figure_size))
plt.subplot(1,3,1),plt.imshow(color_image), plt.title("Imagen original")
plt.subplot(1,3,2),plt.imshow(sharpened), plt.title("Sharpened")

plt.show()
```
